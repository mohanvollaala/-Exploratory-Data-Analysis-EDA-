```python
#matplotlib
#Histogram of purchase frequency
plt.hist(df['Purchase_Frequency'], bins=20, color='skyblue')
plt.title("Purchase Frequency Distribution")
plt.xlabel("Frequency")
plt.ylabel("Count")
plt.show()
```

```python
#Histogram of average order value
plt.hist(df['Average_Order_Value'], bins=15, color='orange')
plt.title("Average Order Value Distribution")
plt.xlabel("Order Value")
plt.ylabel("Count")
plt.show()
```

```python
#boxplot fortime between purchases
plt.boxplot(df['Time_Between_Purchases'])
plt.title("Time Between Purchases Boxplot")
plt.ylabel("Days")
plt.show()
```

```python
#boxplot for lifetime value
plt.boxplot(df['Lifetime_Value'])
plt.title("Lifetime Value Boxplot")
plt.ylabel("Value")
plt.show()
```

```python
#line plot of average order value
plt.plot(df['Average_Order_Value'], color='purple')
plt.title("Average Order Value Trend")
plt.ylabel("Order Value")
plt.show()
```

```python
# 7. Scatter plot between Average Order Value & Lifetime Value
plt.scatter(df['Average_Order_Value'], df['Lifetime_Value'], alpha=0.5)
plt.title("AOV vs Lifetime Value")
plt.xlabel("AOV")
plt.ylabel("Lifetime Value")
plt.show()
```

```python
# 8. Scatter plot between Purchase Frequency & Lifetime Value
plt.scatter(df['Purchase_Frequency'], df['Lifetime_Value'], color='red')
plt.title("Purchase Frequency vs LTV")
plt.xlabel("Frequency")
plt.ylabel("Lifetime Value")
plt.show()
```

```python
df['Most_Frequent_Category'].value_counts().plot(kind='bar', color='teal')
plt.title("Most Frequent Category Distribution")
plt.show()
```

```python
df['Region'].value_counts().plot(kind='bar', color='brown')
plt.title("Region Distribution")
plt.show()
```

```python
df['Region'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Region Share")
plt.show()
```

```python
# 12. Pie chart for Season
df['Season'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Season Share")
plt.show()
```

```python
plt.hist(df['Lifetime_Value'], bins=25, color='lightgreen')
plt.title("Lifetime Value Distribution")
plt.show()
```

```python
# 16. Horizontal bar chart for Preferred Purchase Times
df['Preferred_Purchase_Times'].value_counts().plot(kind='barh', color='pink')
plt.title("Preferred Purchase Times")
plt.show()
```

```python

# 17. Histogram of Peak Sales Date
plt.hist(df['Peak_Sales_Date'], bins=12, color='cyan')
plt.title("Peak Sales Date Distribution")
plt.show()
```

```python
# 18. Scatter plot between Churn Probability & Lifetime Value
plt.scatter(df['Churn_Probability'], df['Lifetime_Value'], alpha=0.5)
plt.title("Churn vs LTV")
plt.xlabel("Churn Probability")
plt.ylabel("Lifetime Value")
plt.show()
```

```python
# 19. Scatter plot between Purchase Frequency & Average Order Value
plt.scatter(df['Purchase_Frequency'], df['Average_Order_Value'], alpha=0.5)
plt.title("Frequency vs AOV")
plt.xlabel("Frequency")
plt.ylabel("AOV")
plt.show()
```
